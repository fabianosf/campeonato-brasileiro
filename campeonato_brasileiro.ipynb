```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get('https://www.cbf.com.br/futebol-brasileiro/competicoes/campeonato-brasileiro-serie-a/2019')
```

```python
if req.status_code == 200:
    print('Ok')
```

```python
content = req.content
```

```python
soup = BeautifulSoup(content, 'html.parser')
table = soup.find_all(name='table')
```

```python
table_str = str(table)
df = pd.read_html(table_str)[0]
```

```python
df
```

```python
df.drop(columns=['Próx','Recentes'], inplace=True)
```

```python
df
```

```python
df.to_csv("tabela_campeonato.csv", sep ='\t')
```
